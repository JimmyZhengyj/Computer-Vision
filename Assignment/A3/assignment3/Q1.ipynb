```python
def LOG(x,y, delta):
    A = (x**2 + y**2)/(2 * delta**2) - 1
    # print("X", x)
    # print("y", y)
    # print("what", -(x**2 + y**2)/(2 * delta**2))
    B = math.exp(-(x**2 + y**2)/(2 * delta**2))
    C = (math.pi * delta**4)
    # print(A)
    # print(B)
    # print(C)
    return ((x**2 + y**2)/(2 * delta**2) - 1)* math.exp(-(x**2 + y**2)/(2 * delta**2)) /(math.pi * delta**4)

def Getconvolution(img, delta):
    m,n = img.shape
    conv = 0
    for i in range(m):
        for j in range(n):
            # print(LOG(i-int(m/2),j-int(n/2),delta))
            conv += img[i][j] * LOG(i-int(m/2),j-int(n/2),delta)

    return conv
```

```python
size = img.shape[0]
result = []
# for delta in [0.1, 1,10,100,1000,10000,100000]:
for delta in range(10,100,10):
    conv = Getconvolution(img, delta)
    print(conv)
    result.append(conv)
    
```

```python
plt.plot(range(10,100,10), result)
plt.xlabel("delta value")
plt.ylabel("convolution")
plt.show()
```
